# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

theme = ["#DEF5E5", "#459E97", "#4B9ABB","#0A4361","#E68193",]
```

```python
data_netflix=pd.read_csv("netflix_titles.csv");
data_disney=pd.read_csv("disney_plus_titles.csv");
```

```python
data_netflix.columns
```

```python
data_disney.columns
```

# Load and clean the data

```python
data_net_clean = data_netflix.drop(['show_id','cast','description','listed_in'], axis = 1)
data_net_clean
```

```python
data_net_clean.shape
```

```python
data_net_clean.columns
```

```python
data_disney_clean = data_disney.drop(['show_id','cast','description','listed_in'], axis = 1)

data_disney_clean
```

```python
data_disney_clean.shape
```

```python
data_disney_clean.columns
```

# Analyzing the data

<p style = "font-size: 117%;" > Amount of titles that have on each one. this shows how many options each platform makes available to subscribers. The bigger, obviously, the more choice the subscriber will have.</p> 

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 9))
fig.suptitle(' Content Comparision ', size = 26, color = theme[3], weight='bold')
axs = [ax1, ax2]

#figure1
fig1 = sns.countplot(x="type",data=data_net_clean,ax=ax1,palette='husl')
fig1.set_ylabel("Count",fontsize=10, weight='bold')
fig1.set_xlabel("Type",fontsize=10,weight='bold')
ax1.set_title(f'Netflix have: {data_net_clean.type.value_counts().sum()}', size = 14, color = theme[3], weight='bold', pad=20)


#figure2

fig2=sns.countplot(x="type", data=data_disney_clean,ax=ax2,palette='husl')
fig2.set_ylabel("Count",fontsize=10, weight='bold')
fig2.set_xlabel("Type",fontsize=10,weight='bold')
ax2.set_title(f'Disney have: {data_disney_clean.type.value_counts().sum()}', size = 14, color = theme[3], weight='bold', pad=20)

plt.subplots_adjust(left=None, bottom=None, right=None, top=0.8, wspace=0.4, hspace=None);
plt.show()
```

<p style="font-size:117%;" > Results: In the amount of movies and TV shows netflix wins. But we must remember that netflix is much "older" than disney+ and that this data is a little older...  

```python
data_disney_clean['Platform']= np.zeros(1450)
data_net_clean['Platform'] = np.ones(8807)
```

```python
#naming new merge dataset as net_dis
frames = [data_disney_clean,data_net_clean]
net_dis = pd.concat(frames)
net_dis.head(n=-10)
```

```python
plt.figure(figsize=(8,6))
plot = sns.countplot(x='type',data=net_dis,hue='Platform',palette='husl')
plot.set_ylabel("Count",fontsize=10, weight='bold')
plot.set_xlabel("Type",fontsize=10,weight='bold')
plt.legend(title='Platform', loc='upper right', labels=['Disney+', 'Netflix'])
plot.figure.suptitle(' Comparision in single plot ',y=1, size = 20, color = theme[3], weight='bold');
```
